# 创建正弦曲线图

创建数据 x 和 y：np.linspace 用于创建100个等间隔的x值，然后通过 np.sin 计算对应的正弦值。

创建图形和轴：plt.subplots 用于创建一个包含图形和轴的对象。

在轴上绘制折线图：ax.plot 用于在轴上绘制折线图，这里是 x 对应于 y 的正弦曲线。

添加标题和标签：ax.set_title、ax.set_xlabel 和 ax.set_ylabel 用于设置图形的标题以及x和y轴的标签。

添加图例：ax.legend 用于添加图例，以便解释曲线的含义。

显示图形：plt.show 用于显示绘制的图形。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一些示例数据
x = np.linspace(0, 10, 100)  # 创建一个包含100个点的从0到10的等间隔数据
y = np.sin(x)               # 计算对应于x的正弦值

# 创建一个图形和轴
fig, ax = plt.subplots()

# 在轴上绘制折线图
ax.plot(x, y, label='sin(x)')

# 添加标题和标签
ax.set_title('Sinusoidal Curve')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')

# 添加图例
ax.legend()

# 显示图形
plt.show()
```

# 散点图

创建数据 x、y、colors 和 sizes：np.random.rand 用于生成随机的x、y坐标、颜色值和大小。

创建图形和轴：plt.subplots 用于创建一个包含图形和轴的对象。

在轴上绘制散点图：ax.scatter 用于在轴上绘制散点图，c 参数用于设置点的颜色，s 参数用于设置点的大小，alpha 参数用于设置透明度，cmap 参数用于设置颜色映射。

添加标题和标签：ax.set_title、ax.set_xlabel 和 ax.set_ylabel 用于设置图形的标题以及x和y轴的标签。

显示颜色条：plt.colorbar 用于显示颜色条，这样可以解释颜色与数值之间的关系。

显示图形：plt.show 用于显示绘制的图形。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一些示例数据
x = np.random.rand(50)  # 创建50个随机x值
y = np.random.rand(50)  # 创建50个随机y值
colors = np.random.rand(50)  # 创建50个随机颜色值，用于设置点的颜色
sizes = 100 * np.random.rand(50)  # 创建50个随机大小，用于设置点的大小

# 创建一个图形和轴
fig, ax = plt.subplots()

# 在轴上绘制散点图
ax.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis')

# 添加标题和标签
ax.set_title('Scatter Plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')

# 显示颜色条
cbar = plt.colorbar(ax.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis'))
cbar.set_label('Color')

# 显示图形
plt.show()
```
