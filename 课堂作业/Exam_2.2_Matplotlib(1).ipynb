# 题目1：散点图

题目：

你有一组城市的数据，包括城市的名称、经度、纬度和人口数量。你希望使用Matplotlib创建一个散点图，将城市的经度和纬度可视化出来，同时用点的大小表示城市的人口数量。

请编写一个Python脚本，完成以下任务：

创建一个包含至少10个城市数据的NumPy数组或Python字典，每个数据包括城市的名称、经度、纬度和人口数量。
使用Matplotlib创建一个散点图，横坐标为城市的经度，纵坐标为城市的纬度，点的大小表示城市的人口数量。确保点的颜色相同。
在图表上添加合适的标题、横坐标标签和纵坐标标签，以使图表清晰易懂。

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建包含城市数据的字典
city_data = {
    'city_name': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'longitude': [-74.006, -118.243, -87.629, -95.369, -112.074, -75.165, -98.493, -117.161, -96.796, -121.886],
    'latitude': [40.712, 34.052, 41.878, 29.760, 33.448, 39.952, 29.424, 32.715, 32.776, 37.338],
    'population': [8398748, 3980808, 2693976, 2320268, 1680992, 1584064, 1547253, 1423851, 1343573, 1035317]
}

size = [a * 0.00025 for a in city_data['population']]

colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k', 'purple', 'orange', 'brown']

plt.scatter(city_data['longitude'], city_data['latitude'], size, colors)
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('city_data')
plt.xlim(-130, -60)
plt.ylim(28, 45)

for i, city in enumerate(city_data['city_name']):
    plt.text(city_data['longitude'][i], city_data['latitude'][i], city, fontsize=12, ha='right')
    
plt.show()
```

# 题目2：折线图

假设你正在研究某城市的月平均气温变化，并收集了一年内的每个月的数据。你希望使用Matplotlib创建一个折线图，将月份（1月至12月）与对应的平均气温可视化出来。

请编写一个Python脚本，完成以下任务：

创建两个包含月份（1至12）和平均气温数据的NumPy数组或Python列表。
使用Matplotlib创建一个折线图，横坐标为月份，纵坐标为平均气温。
添加合适的标题、横坐标标签和纵坐标标签，以使图表清晰易懂。
设置折线图的线型、线条颜色和标记样式（例如，实线、蓝色、圆点标记）。

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建包含月份和平均气温数据的列表
months = ['1月', '2月', '3月', '4月', '5月', '6月', '7月', '8月', '9月', '10月', '11月', '12月']
avg_temperatures = [10, 12, 15, 18, 22, 26, 30, 30, 27, 22, 16, 12]
```

```python
# 请输入你的程序
X = range(len(months))
Y = avg_temperatures
plt.plot(X, Y, marker = 'o', linestyle = '-', markerfacecolor='r', markersize=8)
plt.xlabel('month')
plt.ylabel('avg_temperatures')
plt.title('data')
plt.show()
```
