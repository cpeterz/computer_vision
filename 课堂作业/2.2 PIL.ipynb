# 图像读取、保存与显示

```python
from PIL import Image
from IPython.display import display
im = Image.open("../传统视觉/人脸识别/face_picture/face_picture1.jpg")
# im.show()
display(im)
```

# 几何变换

## 尺度缩放

```python
im_resize = im.resize((512, 512))
display(im_resize)
```

# 图像旋转

```python
im_rot = im_resize.rotate(45) 
display(im_rot)
```

# 模式转换

```python
im_gray = im_resize.convert("L")
display(im_gray)
```

# 图像模糊

均值滤波

```python
from PIL import ImageFilter
im_blur = im_resize.filter(ImageFilter.BLUR)
display(im_blur)
```

# 轮廓检测

```python
image_contour = im_resize.filter(ImageFilter.CONTOUR)

display(image_contour)
```

# 边缘检测

```python
image_edge = im_resize.filter(ImageFilter.FIND_EDGES)
display(image_edge)
```
